```python
## Import the frequency values from the text files you previously downloaded
from numpy import loadtxt

fr = loadtxt("Frequency.txt", comments="#", delimiter=",")
```

```python
## Import the gain values from the 5 text files you previously downloaded, names Gain_data1, Gain_data2, etc
from numpy import loadtxt

Gain_data1 = loadtxt("Gain_data1.txt", comments="#", delimiter=",")
Gain_data2 = loadtxt("Gain_data2.txt", comments="#", delimiter=",")
Gain_data3 = loadtxt("Gain_data3.txt", comments="#", delimiter=",")
Gain_data4 = loadtxt("Gain_data4.txt", comments="#", delimiter=",")
Gain_data5 = loadtxt("Gain_data5.txt", comments="#", delimiter=",")
```

2. Calculate the average of the 5 gains measured at each set freqeuncy. [5 points]

```python
## Show your calculations here
GainAvg = np.zeros((801,))
i = 0

while i <= 800:
    GainAvg[i] = (Gain_data1[i]+Gain_data2[i]+Gain_data3[i]+Gain_data4[i]+Gain_data5[i])/5
    i += 1
    
```

3. Calculate the standard deviation for the gain measured at each frequency. Your answer should be an array containing 801 elements. [5 points]

```python
## Show your calculations here
Gains = np.zeros((801,))
Gain_F = np.zeros((5,))
i = 0

while i <= 800:
    Gain_F[0] = Gain_data1[i]
    Gain_F[1] = Gain_data2[i]
    Gain_F[2] = Gain_data3[i]
    Gain_F[3] = Gain_data4[i]
    Gain_F[4] = Gain_data5[i]
    a = np.std(Gain_F, axis = 0)
    Gains[i] = a
    i += 1
```

4. Make a plot showing each set frequency on the x-axis and the average gain on the y-axis. Do not worry about error bars. [4 points]

```python
## Display your plot here. Hint: use plt.plot(y,x)
y =  GainAvg
x = fr

plt.title("Plot 1: Average Gain vs Frequency")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Average Gain")
plt.plot(x,y)
```

5. Qualitatively describe the shape of this plot. What happens at low frequencies? What happens at high frequencies? [4 points]

Analyzing plot 1, we can see that at low frequencies, the average gain is very large. Inversely, as the frequency increases, the average gain decreases following an L curve shape.

21. Display your experimental data below in Python.
    <br>a. Display the recorded data for Frequency, $v_{in}$, and $v_{out}$ in the first three columns.
    <br>b. Calcutate Gain in the last column.
    <br>*Recall that* \begin{equation*}Gain (dB) = 20log_{10} \frac{v_{out}}{v_{in}} \end{equation*}

```python
## Display your calculations and results here
#display values as arrays
Frequency = np.array([50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900, 2000, 2500, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000])
Vin = np.array([2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2, 2])
Vout = np.array([1.34, 1.8, 5.2, 7.4, 9.4, 11, 12.2, 13.6, 14.4, 15.2, 15.6, 16.2, 16.6, 17.2, 17.4, 17.6, 17.8, 18, 18.2, 18.2, 18.4, 18.6, 19, 19.2, 19.4, 19.6, 19.6, 19.4, 19.2, 19.2])

#convert first three arrays into columns
Frequency_col = Frequency[:,None]
Vin_col = Vin[:,None]
Vout_col = Vout[:,None]

#now calculate gain
V_col = np.divide (Vout_col, Vin_col)
Gain = np.log10(V_col)
Gain_col = np.multiply(20, Gain)

#display all data in columns
print('   Frequency(Hz)   Vin(V)          Vout(V)         Gain(dB)')
print((np.hstack((Frequency_col,Vin_col,Vout_col, Gain_col))))
```
